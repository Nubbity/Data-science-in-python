```python
import pandas as pd
import numpy as np
from sklearn import svm

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
receipes = pd.read_csv(r"C:\Users\hynni\Downloads\recipes_muffins_cupcakes.csv")
receipes
```

```python
sns.lmplot(x = 'Sugar', y = 'Butter', data = receipes, hue = 'Type', fit_reg = False)
```

```python
sugar_butter = receipes[['Sugar', 'Butter']]
type_label = np.where(receipes['Type'] == 'Muffin',0,1)
```

```python
model = svm.SVC(kernel = 'linear')
model.fit(sugar_butter, type_label)
```

```python
print(model.coef_)
```

```python
xx = np.linspace(5,30)
yy = -(model.coef_[0][0]/model.coef_[0][1])*xx -(model.intercept_[0]/model.coef_[0][1])
```

```python
model.support_vectors_
```

```python
w = model.coef_[0]
a = -w[0]/w[1]##DIDNT HAVE TIME TO SEE HOW THIS WAS CALCULATED
b = model.support_vectors_[0]
yy_down = a * xx +(b[1]- a*b[0])

c = model.support_vectors_[-1]
yy_up = a * xx +(c[1]- a*c[0])

sns.lmplot(x = 'Sugar', y = 'Butter', data = receipes, hue = 'Type', fit_reg = False)
plt.plot(xx, yy, color='black')
plt.plot(xx, yy_up, 'k--')
plt.plot(xx, yy_down, 'k--')
```

```python
def muffin_or_cupcake(butter, sugar):
    if(model.predict([[butter, sugar]]) == 0):
        print('You are looking at a muffin recipe')
    else:
        print("You are looking at a cupcake recipe")
```

```python
muffin_or_cupcake(12,12)
```

```python
sugar_butter_original = receipes[['Sugar', 'Butter']]
type_label = np.where(receipes['Type'] == 'Muffin',0,1)

model = svm.SVC(kernel = 'linear', C=1)
model.fit(sugar_butter_original, type_label)

w = model.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(5,30)
yy = -(model.coef_[0][0]/model.coef_[0][1])*xx -(model.intercept_[0]/model.coef_[0][1])

b = model.support_vectors_[0]
yy_down = a * xx +(b[1]- a*b[0])

c = model.support_vectors_[-1]
yy_up = a * xx +(c[1]- a*c[0])

sns.lmplot(x = 'Sugar', y = 'Butter', data = receipes, hue = 'Type', fit_reg = False)
plt.plot(xx, yy, color='black')
plt.plot(xx, yy_up, 'k--')
plt.plot(xx, yy_down, 'k--')


```
